```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
df = pd.read_excel("data.xlsx")
print(df)
```

```python
# to plot the graph
df.plot(x="Year", y="salary growth", kind="line")
plt.show()
```

```python
# to plot the graph
df.plot(x="Year", y="pce yoy", kind="line", color = "red")
plt.show()
```

```python
# to plot the graph
df.plot(x="Year", y="cpi yoy", kind="line", color = "green")
plt.show()
```

```python
# to plot the graph
df.plot(x="Year", y="ppi yoy", kind="line", color = "orange")
plt.show()
```

```python
# Plotting the Data

ax = plt.gca()

df.plot(kind='line',x="Year", y="salary growth",ax=ax)
df.plot(kind='line',x="Year", y="pce yoy", color='red', ax=ax)

plt.show()
```
