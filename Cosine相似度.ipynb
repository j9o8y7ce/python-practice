## 改完Function 寫法

```python
def cos_silimarity(data1, data2, key):
    data1_data2 = pd.merge(data1, data2, on = key).drop(key, 1)
    data1_data2['group'] = data1_data2['word_x'] + data1_data2['word_y']
    data1_data2 = data1_data2.drop(columns = ['word_x', 'word_y'])
    
    data1_len = len(data1.columns)-1
    data2_len = len(data2.columns)-1
    cos_new = pd.DataFrame()
    cos_new['group'] = data1_data2['group']
    
    for i in range(1, data1_len):
        cos_new['xx' + str(i)] = data1_data2['x' + str(i) + '_x'] ** 2
        for j in range(1, data2_len):
            cos_new['x' + str(i) + 'y' + str(i)] = data1_data2['x' + str(i) + '_x'] * data1_data2['x' + str(i) + '_y']
            cos_new['yy' + str(j)] = data1_data2['x' + str(j) + '_y'] ** 2
#        print(cos_si_new)

###########################################
# columns name

    
    cols = ['group']
    for i in range(1, data1_len):
        cols.append('xx' + str(i))
    for i in range(1, data2_len):
        cols.append('yy' + str(i))
    for i in range(1, data1_len):
        cols.append('x' + str(i) + 'y' + str(i))
        
# 更換columns order
    cos_new = cos_new.reindex(columns = cols)

###########################################
# 產出 square root(sum_xx), square root(sum_yy), sum(xiyi), cos_similarity
    cos_new['sr_sum_xx'] = cos_new.iloc[:, 1:51].sum(axis = 1) ** .5
    cos_new['sr_sum_yy'] = cos_new.iloc[:, 51:101].sum(axis = 1) ** .5
    cos_new['sum_xy'] = cos_new.iloc[:, 101:151].sum(axis = 1)
    cos_new['cos_similarity'] = cos_new['sum_xy'] / (cos_new['sr_sum_xx'] * cos_new['sr_sum_yy'])
    
# 排序
    cos_new_sort = cos_new.sort_values(by = 'cos_similarity', ascending = False)    
    
# 產出最接近1的組合與值
    answer = ['The most similar group is : '+ cos_new_sort['group'].iloc[0]\
              + ' and value is '+ str(cos_new_sort['cos_similarity'].iloc[0])]
    
    return answer
```
